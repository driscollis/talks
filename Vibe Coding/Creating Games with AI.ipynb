## What is Vibe Coding?

Using AI to write most or all of your code

## How is Vibe Coding Useful?

- Faster coding
- Quicker learning
- Can be good for test writing

## Vibe Coding Downsides

- You don't learn as well
- Mistakes in the code
- Debugging is harder since you didn't write it

## Vibe Coding a Game in Python

## Common Problems

- Code might not work at all (Ex. Uses variables that were not initialized)
- Code partially works (Ex. Controls are only partially implemented)
- Game logic is wrong (Ex. bad collision detection, can't win)
- Doesn't create what you told it to at all (no maze, score is sent to stdout, etc)

## Did it Work At All?

- Some games worked correctly (Tic-Tac-Toe, Connect Four)
- Some games had the right logic, but not the right UI (Minesweeper)
- Some games had the right UI but could not be won (Lune Lander)

## Other Gotchas

- No sprites
- No sound effect / music

## AI Can Fix Some of This

- You can generate sprites - Open Art AI
- You can create music
